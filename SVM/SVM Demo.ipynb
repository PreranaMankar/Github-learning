```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.arange(-5, 9)
data
```

```python
target = np.array([0,1,0]).repeat([5, 5,4])
target
```

```python
X = data, target
```

```python
X
```

```python
data_sq = data ** 2
data_sq
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(zip(data, data_sq), columns = ["X", "X2"])
```

```python
df["y"] = target
df
```

```python
plt.scatter(df["X"][df["y"]==0], df["X2"][df["y"]==0])
plt.scatter(df["X"][df["y"]==1], df["X2"][df["y"]==1])
plt.show()
```
