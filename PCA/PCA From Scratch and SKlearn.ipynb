```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2.5, 0.5, 2.2, 1.9, 3.1, 2.3, 2])
y = np.array([2.4, 0.7, 2.9, 2.2, 3.0, 2.7, 1.6])
```

```python
cov_xy = np.cov(x,y)
```

```python
cov_xy
```

```python
eig_val, eig_vec =  np.linalg.eig(cov_xy)
```

```python
eig_val
```

```python
eig_vec
```

```python
data = np.append(x.reshape(-1,1),y.reshape(-1,1),axis=1)
```

```python
data
```

```python
pc = np.dot(data, eig_vec[1].T)
```
