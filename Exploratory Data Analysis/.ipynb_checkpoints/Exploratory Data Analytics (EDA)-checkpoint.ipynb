```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("googleplaystore.csv")
```

```python
df = data.copy()
df.head()
```

```python
df.info()
```

```python
df["Reviews"].astype("int64")
```

```python
df["Reviews"].replace("3.0M", "3000000", inplace=True)
```

```python
df["Reviews"] = df["Reviews"].astype("int64")
```

```python
df["Reviews"].dtype
```

```python
df["Size"].value_counts()
```

```python
df = df[df["Size"]!="Varies with device"]
```

```python
df.info()
```

```python
df = df[df["Size"]!='1,000+']
```

```python
dummy_size = []

for i in df["Size"]:
    if i[-1]=="M":
        dummy_size.append(float(i.replace("M", ""))*1024)
    elif i[-1]=="k":
        dummy_size.append(float(i.replace("k", "")))
    else:
        dummy_size.append(float(i))
```

```python
dummy_size
```

```python
df["Size"] = dummy_size
```

```python
df.head()
```

```python
df["Size"] = df["Size"].astype("int64")
```

```python
df.head()
```

```python
df["Installs"] = df["Installs"].str.replace(",","")
df["Installs"] = df["Installs"].str.replace("+","")
```

```python
df["Installs"] = df["Installs"].astype("int64")
```

```python
df["Type"].value_counts()
```

```python
df["Price"].value_counts()
```

```python
df["Price"] = df["Price"].str.replace("$","").astype("float")
```

```python
df["Content Rating"].value_counts()
```

```python
df["Genres"].value_counts()
```

```python
df["Last Updated"] = df["Last Updated"].astype("datetime64[ns]")
```

```python
df.drop(["Genres", "Current Ver", "Android Ver"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df["Category"].value_counts()
```

```python
df["Last Updated"].min()
```

```python
df["Last Updated"].max()
```

```python
df["Reviews"] = df["Reviews"].astype("int64")
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df["Rating"].value_counts().describe()
```

```python
df["Rating"].value_counts()
```

```python
df["Rating"].describe()
```

```python
df["Rating"] = df["Rating"].fillna(3.7)
```

```python
df.info()
```

```python
df = df.reset_index(drop=True)
```

```python
dummy_df = df[["Type", "Price"]]
```

```python
dummy_df.groupby("Type").sum()
```
