```python
def calculate_entropy(pos_spl, neg_spl):
    total = pos_spl + neg_spl
    return -(pos_spl/total)* np.log2(pos_spl/total) - (neg_spl/total)* np.log2(neg_spl/total)
```

```python
overall_entropy = calculate_entropy(9, 5)
overall_entropy
```

```python
sunny_entropy = calculate_entropy(2, 3)
sunny_entropy
```

```python
overcast_entropy = calculate_entropy(4, 0)
overcast_entropy
```

```python
rain_entropy = calculate_entropy(3, 2)
rain_entropy
```

```python
weighted_entropy_outlook = ((5/14)*sunny_entropy) + ((4/14)*0) + ((5/14)*rain_entropy)
weighted_entropy_outlook 
```

```python
ig_outlook = overall_entropy - weighted_entropy_outlook
ig_outlook
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
X = iris.data
y = iris.target
```

```python
X[:3]
```

```python
y[:3]
```

```python
dtc.fit(X,y)
```

```python
from sklearn import tree
```

```python
iris.feature_names
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data["target"] = iris.target
data.head()
```

```python
plt.figure(figsize=(20, 16))
tree.plot_tree(dtc)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y, dtc.predict(X))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=24)
```

```python
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
plt.figure(figsize=(20, 16))
tree.plot_tree(dtc)
plt.show()
```

```python
dtc2 = DecisionTreeClassifier(max_depth=2)
```

```python
dtc2.fit(X_train, y_train)
```

```python
y_pred2 = dtc2.predict(X_test)
```

```python
accuracy_score(y_test, y_pred2)
```

```python
tree.plot_tree(dtc2)
plt.show()
```

```python
# GridSearchCV - Used to find the best parameters from hyperparameters of any algorithm
# This will work on every combination of all the given hyperparameters
# Then will return the best from them 
# This will take the below parameters
# estimator - Model, for ex.Decision Tree, Logistic Regression
# param_grid - Dictionary of parameters of given estimator
# scoring - To calculate score ---- Accuracy, Precision, Recall, f1
# n_jobs - Used to define how many cores of the processor will be used for computation, -1 for all
# cv - cross-validation, how many times data  will be cross-checked 
# verbose - showing how many combinations will be computed
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = DecisionTreeClassifier()
```

```python
params = {"criterion": ["entropy", "gini"],
         "max_depth": [1, 2, 3, 4, 5, 6, 7],
         "max_features": [1, 2, 3, 4]}
```

```python
gsc = GridSearchCV(estimator=model, param_grid=params, n_jobs=-1, cv = 10, verbose=True)
```

```python
gsc.fit(X_train, y_train)
```

```python
gsc.best_params_
```

```python
gsc.best_score_
```

```python
model = DecisionTreeClassifier(criterion="entropy", max_depth=5, max_features=3)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred2 = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred2)
```

```python
plt.figure(figsize=(20, 16))
tree.plot_tree(model)
plt.show()
```
